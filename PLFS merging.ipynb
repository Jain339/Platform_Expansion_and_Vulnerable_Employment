```python
import pandas as pd

files = [
    "plfs_2018_19.csv",
    "plfs_2019_20.csv",
    "plfs_2020_21.csv",
    "plfs_2021_22.csv",
    "plfs_2022_23.csv",
    "plfs_2023_24.csv"
]

agg_dfs = [pd.read_csv(f) for f in files]

final_agg = pd.concat(agg_dfs, ignore_index=True)

# Remove state code 26 (stored as string)
final_agg = final_agg[final_agg["state"].astype(str).str.strip() != "26"].copy()
# drop state code column
final_agg = final_agg.drop(columns=["state_ut_code"], errors="ignore")

# rename sector_ru -> region
final_agg = final_agg.rename(columns={"sector_ru": "region"})
```

```python
final_agg.to_csv(
    "plfs_2018_2024.csv",
    index=False
)
```
